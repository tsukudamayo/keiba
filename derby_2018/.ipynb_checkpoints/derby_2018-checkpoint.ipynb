```python
# race table
race_table = pd.read_html(
    'http://race.netkeiba.com/?pid=race_old&id=c201805021210',
    header=0
)
race_table = race_table[0]
race_table = race_table.drop(0, axis=0)
race_table = race_table.drop(1, axis=0)
race_table
```
