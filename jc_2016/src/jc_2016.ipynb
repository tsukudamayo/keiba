```python
# 過去10年分の結果をCSVファイルに出力する
for i in range(2015, 2005, -1):
    if i >= 2014:
        race_data = pd.io.html.read_html('http://db.sp.netkeiba.com/v0/?pid=race_detail&id=' + str(i) + '05050911')
        df = pd.DataFrame(race_data[0])
        df.to_csv('jc_' + str(i) + '.csv', header=None)
        print(race_data[0])
    elif 2012 <= i <= 2013:
        race_data = pd.io.html.read_html('http://db.sp.netkeiba.com/v0/?pid=race_detail&id=' + str(i) + '05050811')
        df = pd.DataFrame(race_data[0])
        df.to_csv('jc_' + str(i) + '.csv', header=None)
        print(race_data[0])
    elif i < 2012:
        race_data = pd.io.html.read_html('http://db.sp.netkeiba.com/v0/?pid=race_detail&id=' + str(i) + '05050810')
        df = pd.DataFrame(race_data[0])
        df.to_csv('jc_' + str(i) + '.csv', header=None)
        print(race_data[0])
```
